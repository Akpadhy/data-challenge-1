### Load data

```python
%matplotlib inline  

import pandas as pd
from datetime import datetime, timedelta
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import numpy as np

df = pd.read_csv('employee_retention_data.csv')
df.head()
```

```python
df.describe()
```

```python
from datetime import datetime
date_format = "%Y/%m/%d"
a = datetime.strptime('2011/01/24', date_format)
b = datetime.strptime('2015/12/13', date_format)
delta = b - a
days = delta.days + 1
print('Total days: ' + str(days)) 

num_company = len(df.company_id.drop_duplicates())
print('Total number of companies: ' + str(num_company)) 
```

```python
# Compute headcount
headcount = np.zeros((num_company, days))

for index, row in df.iterrows():
    date_start = datetime.strptime(row['join_date'], '%Y-%m-%d')
    ind_start = (date_start - a).days
    if type(row['quit_date']) is float:
        ind_end = days
    else:
        date_end = datetime.strptime(row['quit_date'], '%Y-%m-%d')
        ind_end = (date_end - a).days
    headcount[int(row['company_id']) - 1, ind_start : ind_end] += 1   

# Compute dates    
dates = [(a + timedelta(days=i)).strftime("%Y-%m-%d") for i in range(days)]

# Full list
result = []
for i in range(num_company):
    for j in range(days):
        result.append([dates[j], int(headcount[i, j]),  i + 1])

# Convert to df        
df = pd.DataFrame(result, columns=['day', 'employee_headcount', 'company_id'])        
```
